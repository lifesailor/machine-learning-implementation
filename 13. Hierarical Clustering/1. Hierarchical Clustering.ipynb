```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

# 1. Data

```python
D = 2
s = 4
mu1 = np.array([0, 0])
mu2 = np.array([s, s])
mu3 = np.array([0, s])

N = 900
X = np.zeros((N , D))
X[:300, :]  = np.random.randn(300, D) + mu1
X[300:600, :] = np.random.randn(300, D) + mu2
X[600:, :] = np. random.randn(300, D) + mu3
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

# 2. Cluster

### Dendrogram

```python
Z = linkage(X, 'ward')
print("Z.shape", Z.shape)
```

index 1,2 - index

index 3 - distance

index 4 - sample count

```python
plt.figure(figsize=(15,10))
plt.title("ward")
dendrogram(Z)
plt.show()
```

### Single

```python
plt.figure(figsize=(15,10))
Z = linkage(X, 'single')
plt.title("Single")
dendrogram(Z)
plt.show()
```

### Complete

```python
Z = linkage(X, 'complete')
plt.title("Complete")
dendrogram(Z)
plt.show()
```
