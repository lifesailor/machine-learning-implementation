```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Linear Regression - 1d

```python
X = []
Y = []

for line in open('./data_1d.csv'):
    x, y = line.split(',')
    X.append(float(x))
    Y.append(float(y))
    
X = np.array(X)
Y = np.array(Y)

plt.scatter(X, Y)
```

```python
denominator = X.dot(X) - X.mean() * X.sum()
a = (X.dot(Y) - Y.mean() * X.sum()) / denominator
b = (Y.mean() * X.dot(X) - X.mean() * X.dot(Y)) / denominator
```

```python
Yhat = a*X + b
```

```python
plt.scatter(X, Y)
plt.plot(X, Yhat, c='r')
```

### 2. R Square

```python
d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1)/d2.dot(d2)
print(r2)
```
